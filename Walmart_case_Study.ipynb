```python
#Sample sige of male Customers

mean_purcahse=[]
for sample_size in range(50, 2000):
  sample_mean=Male_customer['Total_Purchase'].sample(sample_size).mean()
  mean_purcahse.append(sample_mean)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.arange(50, 2000),mean_purcahse)
plt.title('Male Mean Purchase Values as sample Size increases')
plt.xlabel('Size of Sample')
plt.ylabel('mean value of Purchase')
plt.show()
```

```python
male_mean=[]
size=Male_customer['Total_Purchase'].shape[0]
for bootstrp in range(1000):
  sample_mean=Male_customer['Total_Purchase'].sample(size, replace=True).mean()
  male_mean.append(sample_mean)
```

```python
plt.figure(figsize=(10,8))
sns.histplot(male_mean, kde=True, bins=100, fill=True)
male_ll_90 = np.percentile(male_mean, 5)
    # calculating the lower limit of the 90% confidence interval
male_ul_90 = np.percentile(male_mean, 95)
    # calculating the upper limit of the 90% confidence interval
plt.axvline(male_ll_90, label = f'male_ll_90 : {round(male_ll_90, 2)}', linestyle = '--')
    # adding a vertical line at the lower limit of the 90% confidence interval
plt.axvline(male_ul_90, label = f'male_ul_90 : {round(male_ul_90, 2)}', linestyle = '--')
    # adding a vertical line at the upper limit of the 90% confidence interval

male_ll_95 = np.percentile(male_mean, 2.5)
male_ul_95 = np.percentile(male_mean, 97.5)
plt.axvline(male_ll_95, label = f'male_ll_95 : {round(male_ll_95, 2)}', linestyle = '--', color = 'g')
plt.axvline(male_ul_95, label = f'male_ul_95 : {round(male_ul_95, 2)}', linestyle = '--', color = 'g')


male_ll_99 = np.percentile(male_mean, 0.5)
male_ul_99 = np.percentile(male_mean, 99.5)
plt.axvline(male_ll_99, label = f'male_ll_99 : {round(male_ll_99, 2)}', linestyle = '--', color = 'r')
plt.axvline(male_ul_99, label = f'male_ul_99 : {round(male_ul_99, 2)}', linestyle = '--', color = 'r')

plt.legend()     # displaying a legend for the plotted lines.
plt.plot()       # displaying the plot.
```

from above bootstraping method we have been able to estimate the confidence interval for the total purchase made by each male customer on Black Friday at Walmart, despite having data for only 2482 male individuals. This data provides us with a approximate range within which the total purchase of each male customer falls in a certain level of confidence.

```python
Single_mean=[]
size=df_single['Total_Purchase'].shape[0]
for bootstrp in range(1000):
  sample_mean=df_single['Total_Purchase'].sample(size, replace=True).mean()
  Single_mean.append(sample_mean)
```

```python
plt.figure(figsize = (10, 6))     # setting the figure size of the plot

sns.histplot(Single_mean, kde = True, bins = 100, fill = True)
#for 90% Confidance Interval
# for lower limit
single_lwl_90=np.percentile(Single_mean,5)
#for upper linit
single_upl_90=np.percentile(Single_mean,95)

plt.axvline(single_lwl_90, label=f'single_lwl_90 {round(single_lwl_90,2)}',linestyle='--')
plt.axvline(single_upl_90, label=f'single_upl_90{round(single_upl_90,2)}', linestyle='--')

#fro 95% Confidance interval
single_lwl_95=np.percentile(Single_mean,2.5)
single_upl_95=np.percentile(Single_mean,97.5)
plt.axvline(single_lwl_95, label=f'single_lwl_95{ round(single_lwl_95,2)}',linestyle='--', color='g')
plt.axvline(single_upl_95, label=f'single_upl_95{ round(single_upl_95,2)}',linestyle='--', color='g')
#for 99% confidance Interval

single_lwl_99=np.percentile(Single_mean,0.5)
single_upl_99=np.percentile(Single_mean,99.5)
plt.axvline(single_lwl_99, label=f'single_lwl_99{ round(single_lwl_99,2)}',linestyle='--', color='r')
plt.axvline(single_upl_99, label=f'single_upl_99{ round(single_upl_99,2)}',linestyle='--', color='r')

plt.legend()
plt.show()
```

Through the bootstrapping method, we have been able to estimate the confidence interval for the total purchase made by each single customer on Black Friday at Walmart, despite having data for only 1989 individuals having single as marital status. This provides us with a reasonable approximation of the range within which the total purchase of each single customer falls, with a certain level of confidence.
